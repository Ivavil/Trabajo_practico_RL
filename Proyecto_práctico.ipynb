### 2.1. Configuración de Memoria y Procesador
Para habilitar el *Experience Replay* y procesar las secuencias de frames, instanciamos la memoria secuencial y el procesador de Atari.

````python
# Utilizamos SequentialMemory para almacenar la experiencia del agente
memory = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
processor = AtariProcessor()
````

### 2.2. Definición de la Política de Exploración (Propuestas de Mejora)
Aquí definimos cómo el agente equilibrará la exploración y la explotación. Se han propuesto dos mejoras respecto a la política base. Descomentar la política deseada para su evaluación.

````python

policy = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps', value_max=1., value_min=0.1, value_test=0.05, nb_steps=1000000)
````

### 2.3. Configuración y Compilación del Agente DQN
Instanciamos el agente aplicando la *Target Network* (`target_model_update`) para dar estabilidad al aprendizaje. También se propone como mejora probar diferentes optimizadores.

````python
from tensorflow.keras.optimizers import Adam, RMSprop

# El parámetro target_model_update es crucial para la estabilidad
dqn = DQNAgent(model=model, nb_actions=nb_actions, policy=policy, memory=memory,
               processor=processor, nb_steps_warmup=50000, gamma=0.99,
               target_model_update=10000, train_interval=4, delta_clip=1.)

# Compilación base
dqn.compile(Adam(learning_rate=0.00025), metrics=['mae'])
````

````python
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
checkpoint_weights_filename = 'dqn_' + env_name + '_weights_{step}.h5f'
log_filename = 'dqn_{}_log.json'.format(env_name)

callbacks = [ModelIntervalCheckpoint(checkpoint_weights_filename, interval=250000)]
callbacks += [FileLogger(log_filename, interval=100)]

dqn.fit(env, callbacks=callbacks, nb_steps=1000000, log_interval=10000, visualize=False)


dqn.save_weights(weights_filename, overwrite=True)
````

````python
from __future__ import division

from PIL import Image
import numpy as np
import gym

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten, Convolution2D, Permute


from tensorflow.keras.optimizers import Adam

import tensorflow.keras.backend as K

from rl.agents.dqn import DQNAgent
from rl.policy import LinearAnnealedPolicy, BoltzmannQPolicy, EpsGreedyQPolicy
from rl.memory import SequentialMemory
from rl.core import Processor
from rl.callbacks import FileLogger, ModelIntervalCheckpoint
````

#### Configuración base

````python
#Descomentar y adaptar a cada entorno según corresponda
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4
#
#env_name = 'CartPole-v0'
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n
````

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

1. Implementación de la red neuronal

````python
shape = INPUT_SHAPE + (WINDOW_LENGTH,)

model = tf.Sequential([

    tf.layers.Input(shape=shape),

    # 1) Convolución 1
    tf.layers.Conv2D(32, (8, 8), strides=(4, 4), activation='relu'),

    # 2) Convolución 2
    tf.layers.Conv2D(64, (4, 4), strides=(2, 2), activation='relu'),

    # 3) Convolución 3
    tf.layers.Conv2D(64, (3, 3), strides=(1, 1), activation='relu'),

    tf.layers.Flatten(),

    # 4) Capa densa
    tf.layers.Dense(512, activation='relu'),

    # 5) Salida Q(s,a)
    tf.layers.Dense(nb_actions, activation='linear')
])

print(model.summary())

````

2. Implementación de la solución DQN

````python
# Testing part to calculate the mean reward
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
dqn.load_weights(weights_filename)
dqn.test(env, nb_episodes=10, visualize=False)
````
