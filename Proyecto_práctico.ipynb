````python
from __future__ import division

from PIL import Image
import numpy as np
import gym

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten, Convolution2D, Permute


from tensorflow.keras.optimizers import Adam

import tensorflow.keras.backend as K

from rl.agents.dqn import DQNAgent
from rl.policy import LinearAnnealedPolicy, BoltzmannQPolicy, EpsGreedyQPolicy
from rl.memory import SequentialMemory
from rl.core import Processor
from rl.callbacks import FileLogger, ModelIntervalCheckpoint
````

#### Configuración base

````python
#Descomentar y adaptar a cada entorno según corresponda
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4
#
env_name = 'ALE/Enduro-v0'
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n
````

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

1. Implementación de la red neuronal

````python
shape = INPUT_SHAPE + (WINDOW_LENGTH,)

model = tf.Sequential([

    tf.layers.Input(shape=shape),

    # 1) Convolución 1
    tf.layers.Conv2D(32, (8, 8), strides=(4, 4), activation='relu'),

    # 2) Convolución 2
    tf.layers.Conv2D(64, (4, 4), strides=(2, 2), activation='relu'),

    # 3) Convolución 3
    tf.layers.Conv2D(64, (3, 3), strides=(1, 1), activation='relu'),

    tf.layers.Flatten(),

    # 4) Capa densa
    tf.layers.Dense(512, activation='relu'),

    # 5) Salida Q(s,a)
    tf.layers.Dense(nb_actions, activation='linear')
])

print(model.summary())

````

2. Implementación de la solución DQN

````python
# ==========================================
# MODELO DQN BÁSICO (Classic Deep Q-Network)
# ==========================================

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Input, Permute
from tensorflow.keras.optimizers import Adam
import tf_keras.src.saving.serialization_lib as serialization_lib

memory_basic = SequentialMemory(limit=1000000, window_length=WINDOW_LENGTH)
processor_basic = AtariProcessor()

policy_basic = LinearAnnealedPolicy(
    EpsGreedyQPolicy(),
    attr='eps',
    value_max=1.0,
    value_min=0.1,
    value_test=0.05,
    nb_steps=1000000
)

serialization_lib.in_safe_mode = lambda: False
# ---------------------------------------------------------

dqn_basic = DQNAgent(
    model=model,
    nb_actions=nb_actions,
    policy=policy_basic,
    memory=memory_basic,
    processor=processor_basic,
    nb_steps_warmup=10000,
    gamma=0.99,
    target_model_update=10000,
    train_interval=4,
    delta_clip=1.,
    enable_dueling_network=False
)

dqn_basic.compile(Adam(learning_rate=0.00025), metrics=['mae'])
````

````python
# =====================================================================
# ENTRENAMIENTO VARIANTE DQN BÁSICO
# =====================================================================
from rl.callbacks import FileLogger, ModelIntervalCheckpoint

# Nombres de archivos específicos para la variante básica
weights_basic_filename = 'dqn_basic_{}_weights.h5f'.format(env_name)
checkpoint_basic_filename = 'dqn_basic_' + env_name + '_weights_{step}.h5f'
log_basic_filename = 'dqn_basic_{}_log.json'.format(env_name)

nb_steps_training = 10000

# Callbacks para la variante básica
callbacks_basic = [ModelIntervalCheckpoint(checkpoint_basic_filename, interval=250000)]
callbacks_basic += [FileLogger(log_basic_filename, interval=100)]

print("=== Entrenando variante 'base' durante 1,000,000 pasos ===")
dqn_basic.fit(
    env, 
    callbacks=callbacks_basic, 
    nb_steps=nb_steps_training, 
    log_interval=10000, 
    visualize=False
)

# Guardar los pesos finales del DQN Básico
dqn_basic.save_weights(weights_basic_filename, overwrite=True)
print(f"Entrenamiento completado. Pesos guardados en: {weights_basic_filename}")
````

````python
# ========================================
# GRÁFICA DE RENDIMIENTO DESDE TENSORBOARD
# ========================================
import os
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

# 1. Definir la ruta de los logs de TensorBoard de SB3
# Nota: Si el log está en DQN_2 o DQN_3, cámbialo aquí
log_dir_dueling = './dueling_dqn_tensorboard/DQN_1'

try:
    print("Extrayendo métricas...")
    event_acc_dueling = EventAccumulator(log_dir_dueling)
    event_acc_dueling.Reload()

    # 2. Extraer la recompensa media ('rollout/ep_rew_mean' es la etiqueta de SB3)
    if 'rollout/ep_rew_mean' in event_acc_dueling.Tags()['scalars']:

        # Extraemos leyendo las propiedades del objeto ScalarEvent
        events_dueling = event_acc_dueling.Scalars('rollout/ep_rew_mean')
        step_nums_dueling = [event.step for event in events_dueling]
        vals_dueling = [event.value for event in events_dueling]

        # 3. Graficar la evolución
        plt.figure(figsize=(10, 6))

        # Dibujamos la línea del Dueling DQN
        plt.plot(step_nums_dueling, vals_dueling, label='Dueling DQN', color='orange', alpha=0.9, linewidth=2)

        # Configuración visual de la gráfica
        plt.title('Evolución de la Recompensa Media en Enduro (SB3)')
        plt.xlabel('Pasos de Entrenamiento')
        plt.ylabel('Recompensa Media (100 Episodios)')
        plt.legend(loc='lower right')
        plt.grid(True, linestyle='--', alpha=0.6)

        # Mostrar la gráfica
        plt.tight_layout()
        plt.show()
    else:
        print("Error: No se encontró la etiqueta 'rollout/ep_rew_mean' en el log.")

except Exception as e:
    print(f"Error al procesar los logs de TensorBoard: {e}")
````

Como se puede observar en la gráfica anterior, hay una caida en el aumento de la recompensa a partir del step 1.3M aproximadamente. Esta caída se debe principalmente a una combinación de overfitting y la finalización del exploration decay: Por un lado, tras superar ese número de pasos de entrenamiento, las redes encargadas de estimar el valor del estado y la ventaja en la arquitectura Dueling DQN empiezan a sufrir de overfitting ante escenarios y trayectorias muy específicos del juego, perdiendo su capacidad de generalización. Por otro lado, el proceso de exploration decay de la política $\epsilon$-greedy ya ha alcanzado su valor mínimo, lo que obliga al agente a priorizar casi por completo la explotación mediante una greedy policy. Al carecer de la exploración suficiente para corregir sus errores de sobreajuste o descubrir nuevas estrategias de conducción adaptativas, el agente queda atrapado en máximos locales y su recompensa media comienza a decaer de forma progresiva.

3. Justificación de los parámetros seleccionados y de los resultados obtenidos

Los parámetros seleccionados corresponden a una configuración estándar de DQN para entornos Atari. Se eligió una entrada de 84×84 píxeles con 4 frames apilados para reducir la complejidad de la observación y, al mismo tiempo, capturar información temporal del movimiento. La arquitectura convolucional con capas intermedias y una capa densa final permite extraer características espaciales relevantes y estimar el valor Q de cada acción posible.

La política epsilon-greedy con decaimiento lineal se utilizó para favorecer una exploración intensa al inicio del entrenamiento y una explotación progresiva de la experiencia adquirida. Asimismo, se fijaron valores como gamma = 0.99, learning rate = 0.00025 y una memoria de gran capacidad porque son parámetros habituales en DQN y contribuyen a estabilizar el aprendizaje.

En cuanto a los resultados, la comparación entre el DQN base y el Dueling DQN permite evaluar si la mejora arquitectónica aporta ventajas reales. Una mayor recompensa media y una curva más estable indican un mejor desempeño del agente. Sin embargo, los resultados también dependen del número de pasos de entrenamiento, que en este caso es relativamente reducido, por lo que la evaluación debe interpretarse como una prueba preliminar y no como una medida definitiva del rendimiento del modelo.
